```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("D:\\DATASETS\\GOAT.xlsx",parse_dates=['date'])
```

```python
df
```

```python
df=df.resample('M',on='date').mean()
```

```python
df
```

```python
df['temp_max'].plot(figsize=(18,7))
plt.title("Seasonal Pattern  of Maximum Temperature of Adilabad from 2018 to 2022")
plt.xlabel("Date")
plt.ylabel("Maximum Temperature")
plt.legend(['Max temp'])
plt.show()
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 18, 8
decomposition = sm.tsa.seasonal_decompose(df['temp_max'], model='additive')
fig = decomposition.plot()
plt.show()
```

```python
adfuller(df['temp_max'])
```

```python
plot_pacf(df['temp_max'])
plot_acf(df['temp_max'])
plt.show()
```

```python
df['diff12']=df['temp_max'].diff(12)
```

```python
df.head(3)
```

```python
plot_pacf(df['diff12'].dropna())
plot_acf(df['diff12'].dropna())
plt.show()
```

```python
train=df[:len(df)*80//100]
test=df[len(df)*80//100:]
```

```python
model=sm.tsa.statespace.SARIMAX(train['temp_max'],order=(1,0,2),seasonal_order=(1,0,2,12))
```

```python
model_fit=model.fit()
```

```python
predictions=model_fit.predict(start=test.index[0],end=test.index[-1])
```

```python
test['temp_max'].plot(figsize=(18,7),legend=True)
predictions.plot(legend=True)
plt.grid()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
Mean_Squared_Error=np.sqrt(mean_squared_error(test['temp_max'],predictions))
```

```python
Mean_Squared_Error
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
Mean_Absolute_Error=mean_absolute_error(test['temp_max'],predictions)
```

```python
Mean_Absolute_Error
```

```python
arr1=np.array(Mean_Squared_Error)
arr2=np.array(Mean_Absolute_Error)
```

```python
plt.figure(figsize=(10,6))
colors = ['r', 'g']
tick_label=['Mean Sqaured Error','Mean Absolute Error']
plt.bar(['MSE', 'MAE'], [arr1, arr2],tick_label=tick_label,color=colors)
plt.xlabel("Error Type")
plt.ylabel("Error Value")
plt.show()
```

```python
future=pd.DataFrame(pd.date_range("2022-10-01","2023-09-30",freq='MS'),columns=['date'])
```

```python
future.set_index('date',inplace=True)
```

```python
Forecast=model_fit.predict(start=future.index[0],end=future.index[-1])
```

```python
ax=df['temp_max'].plot(label="Actual Past Data")
predictions.plot(ax=ax,label="Prediction on Test data")
Forecast.plot(ax=ax,label="Predicted Future data",figsize=(18,7))
plt.xlabel("Date")
plt.ylabel("Max_Temp")
plt.legend(["Actual Past Data","Prediction on Test data","Predicted Future Data"])
plt.grid()
plt.show()
```
