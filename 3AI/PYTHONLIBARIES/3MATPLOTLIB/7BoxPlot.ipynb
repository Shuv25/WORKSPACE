```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=['Python'],patch_artist=True,showmeans=True,sym='r',boxprops=dict(color='r'),capprops=dict(color='y'))
plt.show()
```

```python
x=[10,20,30,40,50,60]

plt.boxplot(x,vert=False)
plt.show()
```

```python
x=[10,20,30,40,50,60,70,120]
y=[10,20,30,40,50,60,70]
z=[x,y]

plt.boxplot(z,labels=['Python','C++'],patch_artist=True,showmeans=True,sym='r',boxprops=dict(color='r'),capprops=dict(color='y'))
plt.show()
```
