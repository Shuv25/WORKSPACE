```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=list(np.random.randint(10,60,50))
print(x)
```

```python
l=[10,20,30,40,50,60]
plt.hist(x,color='g',bins=l,edgecolor='r',bottom=10)


plt.title("Histogram")
plt.xlabel("Points")
plt.ylabel("Numbers")

plt.show()
```

```python
l=[10,20,30,40,50,60]
plt.hist(x,color='g',bins=l,edgecolor='r',bottom=10,orientation='horizontal')


plt.title("Histogram")
plt.xlabel("Points")
plt.ylabel("Numbers")

plt.show()
```
