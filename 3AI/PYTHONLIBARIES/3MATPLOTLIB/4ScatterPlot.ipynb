```python
import matplotlib.pyplot as plt
```

```python
x=[3,4,5,6,7,8,9]
y=[5,6,8,10,11,12,14]

plt.scatter(x,y,c=['r','g','b','y','r','g','b'],s=[30,9,40,5,8,50,7],marker="*")#c for color and s for size of each data points

plt.title("SatterPlot")
plt.xlabel("Day")
plt.ylabel("No")
plt.show()
```

```python
x=[3,4,5,6,7,8,9]
y=[5,6,8,10,11,12,14]

plt.scatter(x,y,c=[10,3,6,9,5,2,7],s=[30,9,40,5,8,50,7],cmap='viridis')#c for color and s for side of each data points
plt.colorbar()
plt.title("SatterPlot")
plt.xlabel("Day")
plt.ylabel("No")
plt.show()
```
