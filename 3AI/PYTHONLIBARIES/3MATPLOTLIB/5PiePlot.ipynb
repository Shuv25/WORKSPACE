```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=['C','C++','Java','Python']

ex=[0.3,0.0,0.0,0.0]

plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%",shadow=True,radius=1.1)

plt.title("Pie Chart")
plt.legend(loc=2)
plt.show()
```

Dot Pie Chart

```python
plt.pie([1],radius=0.5,colors='r')
plt.show()
```

Donought Chart

```python
x=[10,20,30,40]
y=['C','C++','Java','Python']

plt.pie(x,labels=y,radius=1,shadow=True)
plt.pie([1],colors='w',radius=0.5)

plt.title("Donought Chart")
plt.legend(loc=2)
plt.show()
```
