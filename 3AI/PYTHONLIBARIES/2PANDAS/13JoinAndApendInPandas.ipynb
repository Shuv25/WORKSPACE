```python
import pandas as pd
```

```python
df=pd.DataFrame({'a':[3,4,5,6],'b':[7,8,9,0]})
ndf=pd.DataFrame({'c':[11,12,13,14],'d':[24,25,26,27]})
```

```python
df.join(ndf)
```

```python
p=pd.DataFrame({'a':[3,4,5,6],'b':[7,8,9,0]})
q=pd.DataFrame({'c':[11,12],'d':[24,25]})

p.join(q)
```

```python
q.join(p)
```

```python
q.join(p,how='right')
```

```python
q.join(p,how='outer')
```
