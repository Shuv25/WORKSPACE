```python
import pandas as pd
```

Melt

```python
df=pd.DataFrame({'days':[1,2,3,4,5,6],'english':[33,44,55,66,77,88],'maths':[43,54,65,76,87,98]})
df
```

```python
pd.melt(df)#used for reshaping
```

```python
pd.melt(df,id_vars='english')
```

```python
pd.melt(df,id_vars='days')
```

```python
pd.melt(df,id_vars='days',var_name="Subject")
```

```python
pd.melt(df,id_vars='days',var_name='Subjects',value_name='Marks')
```

Pivot

```python
ndf=pd.DataFrame({'days':[1,2,3,4,5,6],'st_name':['a','b','c','d','e','f'],'english':[33,44,55,66,77,88],'maths':[43,54,65,76,87,98]})
ndf
```

```python
ndf.pivot(index='days',columns='st_name')
```

```python
ndf.pivot(index='days',columns='st_name',values='english')
```

```python
nndf=pd.DataFrame({'days':[1,1,1,1,2,2],'st_name':['a','b','c','d','e','f'],'english':[33,44,55,66,77,88],'maths':[43,54,65,76,87,98]})
nndf
```

```python
nndf.pivot_table(index='st_name',columns='days',aggfunc="mean",margins=True)
```
