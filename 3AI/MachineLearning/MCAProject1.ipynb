```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df['Target'] = encoder.fit_transform(df['Target'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
target_attribute = 'Target'

features_to_plot = [col for col in df.columns if col != target_attribute]

n_rows = len(features_to_plot) // 2 + len(features_to_plot) % 2
n_cols = 2

fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(15, 20))

axes = axes.flatten()

for i, feature in enumerate(features_to_plot):
    sns.countplot(data=df, x=feature, hue=target_attribute, ax=axes[i])
    axes[i].set_title(f"{feature} vs. {target_attribute}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")

if len(features_to_plot) % 2 == 1:
    fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

```python
features = df.drop('Target', axis=1)
target = df['Target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
target_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, confusion_matrix
```

```python
mse = mean_squared_error(y_test, target_pred)
mae = mean_absolute_error(y_test, target_pred)
```

```python
print(f"Mean Squared Error (MSE): {mse}")
print(f"Mean Absolute Error (MAE): {mae}")
```

```python
model.score(X_test,y_test)
```

```python
confusion = confusion_matrix(y_test, target_pred)
```

```python
sns.heatmap(confusion, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
