```python
import matplotlib.pyplot as plt 
```

```python
x=[30,40,50,60,70,80]

y=[50,30,70,40,30,40]
p=[50,30,70,40,60,40]
q=[30,70,40,80,30,40]

l=['areaone','areatwo','areathree']

plt.stackplot(x,y,p,q,labels=l)

plt.legend()
plt.show()
```

```python
x=[30,40,50,60,70,80]

y=[50,30,70,40,30,40]
p=[50,30,70,40,60,40]
q=[30,70,40,80,30,40]

l=['areaone','areatwo','areathree']

plt.stackplot(x,y,p,q,labels=l,baseline='sym')

plt.title("Area Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()
plt.show()
```
