```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[1,2,3,4]

plt.subplot(2,2,1)
plt.plot(x,y,color='r')

plt.subplot(2,2,2)
plt.pie([1],radius=0.5,colors='g')

plt.subplot(2,2,3)
u=[10,20,30,40]
plt.pie(u,radius=0.5)

plt.subplot(2,2,4)
p=['a','b','c','d']
q=[10,20,30,40]
plt.bar(p,q,color=['r','g','y','b'])

plt.savefig('Subplot2.jpg',dpi=1000,transparent=True)
plt.show()
```
