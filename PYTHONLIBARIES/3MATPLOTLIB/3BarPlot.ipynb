```python
import matplotlib.pyplot as plt
```

```python
x=['python','c','c++','java']
y=[85,70,60,83]
c=['r','g','b','y']
plt.xlabel("Langauges",fontsize=11)
plt.ylabel("Points based upon Popularity",fontsize=11)
plt.title("Popularity",fontsize=11)
plt.bar(x,y,color=c,width=0.4,edgecolor='r',linewidth=3,linestyle=":",alpha=1,label="Shuv")#bar widthalpha value used to increase or decrease the intensity of the pic its from 0 to one.
plt.legend()
plt.show()
```

Overlaping Bar Graph

```python
z=[30,40,50,60]
plt.xlabel("Langauges",fontsize=11)
plt.ylabel("Points based upon Popularity",fontsize=11)
plt.title("Popularity",fontsize=11)
plt.bar(x,y,color='r',width=0.4,linewidth=3,linestyle=":",alpha=1,label="Shuv")
plt.bar(x,z,color='y',width=0.4,linewidth=3,linestyle=":",alpha=1,label="NewShuv")
plt.legend()
plt.show()
```

SidebySde Bar

```python
import numpy as np

widthi=0.2
p=np.arange(len(x))
q=[j+widthi for j in p]

plt.xlabel("Langauges",fontsize=11)
plt.ylabel("Points based upon Popularity",fontsize=11)
plt.title("Popularity",fontsize=11)

plt.bar(p,z,color='y',width=widthi,label="New")
plt.bar(q,y,color='r',width=widthi,label="old")

plt.xticks(p+widthi/2,x,rotation=30)#for naming the x axis values 
plt.legend()
plt.show()
```

Horizontally

```python
widthi=0.2
p=np.arange(len(x))
q=[j+widthi for j in p]

plt.ylabel("Langauges",fontsize=11)
plt.xlabel("Points based upon Popularity",fontsize=11)
plt.title("Popularity",fontsize=11)

plt.barh(p,z,color='y',height=widthi,label="New")
plt.barh(q,y,color='r',height=widthi,label="old")

plt.yticks(p+widthi/2,x,rotation=30)#for naming the x axis values 
plt.legend()
plt.show()
```

```python
o=['Dell','Hp','Samsung','Asus','Acer','Mac','Lenovo']
m=[3,4,5,6,7,8,9]
n=[9,8,7,6,5,4,3]

w=0.3

s=np.arange(len(o))
t=[j+w for j in s]

plt.xlabel("Laptops")
plt.ylabel("Popularity")
plt.title("Popularity")

plt.bar(s,m,width=w,color='r',label="Old")
plt.bar(t,n,width=w,color='y',label="New")

plt.xticks(s+w/2,o,rotation=30)
plt.legend()
plt.show()
```
